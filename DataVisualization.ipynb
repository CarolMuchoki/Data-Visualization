```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('mcs_ds_edited_iter_shuffled.csv')
df.head()
```

```python
sns.boxplot (data = df, x = 'anchor_ratio', y = 'iterations')
```

```python
sns.scatterplot(data=df, x="anchor_ratio", y="ale")
```

```python
sns.lineplot(data=df, x ='anchor_ratio', y = 'sd_ale')
```

```python
sns.histplot(data=df, x="iterations", kde=True)
```

```python
data = df.groupby("anchor_ratio")["iterations"].sum()
palette_color = sns.color_palette('bright')
data.plot.pie(colors = palette_color, autopct="%.1f%%")
```

```python
matrix = df.corr().round(2)
sns.heatmap(matrix, annot=True)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
text = ''
stopwords = set(STOPWORDS)
for val in df.iterations:
    val =str (val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        text += " ". join (tokens) + " "
wordcloud = WordCloud(stopwords = stopwords).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
